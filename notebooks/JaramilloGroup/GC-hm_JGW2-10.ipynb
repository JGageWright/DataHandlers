```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from importer_snippets import biologic_mpt_voltammetry
from LinReg import PolyReg
from scipy import interpolate
import os
plt.style.use("JGW-PPT")

plt.rcParams["figure.figsize"] = (8,6)
```

```python
dir_name = r"C:\Users\jgage\OneDrive - Stanford\Jaramillo Group\EC-data\GC-hm_JGW2-10"
technique = 'CVA'

file_list = []
data_frames = {}
for file in os.listdir(dir_name):
    if file.endswith(".mpt"):
        if technique in str(file):
            path = os.path.join(dir_name, file)
            df = biologic_mpt_voltammetry(path)
            data_frames[file] = df
            # print(str(file)[:-4] + '.csv')
            # df.to_csv(str(path)[:-4] + '.csv')
            file_list.append(file)

file_list
```

```python
series_11_13_names = ['1200rpm-2.5mm_JGW2-10-11_04_CVA_C02.mpt',
                '1200rpm-2.5mm_JGW2-10-12_04_CVA_C02.mpt',
                '1200rpm-2.5mm_JGW2-10-13_04_CVA_C02.mpt']

series_14_16_names = ['1200rpm-2.5mm_JGW2-10-14_04_CVA_C02.mpt',
                '1200rpm-2.5mm_JGW2-10-15_04_CVA_C02.mpt',
                '1200rpm-2.5mm_JGW2-10-16_04_CVA_C02.mpt']


series_height_names = ['1200rpm-1.75mm_JGW2-10-19_02_CVA_C02.mpt',
                 '1200rpm-2mm_JGW2-10-18_02_CVA_C02.mpt',
                 '1200rpm-2.25mm_JGW2-10-21_02_CVA_C02.mpt',
                 # '1200rpm-2.5mm_JGW2-10-16_04_CVA_C02.mpt',
                 '1200rpm-2.75mm_JGW2-10-20_02_CVA_C02.mpt']

series_11_13, series_14_16, series_height, just_17 = [], [], [], []

just_17_names = ['1200rpm-2.5mm_JGW2-10-17_02_CVA_C02.mpt']

for i in series_11_13_names:
    series_11_13.append(data_frames[i])

for i in series_14_16_names:
    series_14_16.append(data_frames[i])

for i in series_height_names:
    series_height.append(data_frames[i])

for i in just_17_names:
    just_17.append(data_frames[i])
```

```python
for i in range(len(series_11_13)):
    # Separate forward and reverse scans
    scan_direction_bools = []
    for k in range(len(series_11_13[i]['Ewe/V'])):
        if k == 0:
            scan_direction_bools.append(True)
        elif series_11_13[i]['Ewe/V'][k - 1]  < series_11_13[i]['Ewe/V'][k]:
            scan_direction_bools.append(True)
        else:
            scan_direction_bools.append(False)

    series_11_13[i]['forward'] = scan_direction_bools

labels = ['11', '12', '13']
selected_cycle = 3
ilc_est = []

for i in range(len(series_11_13)):
    fig, ax = plt.subplots()

    # Convert to RHE, current in Amps
    series_11_13[i]['E vs RHE'] = series_11_13[i]['Ewe/V'] + 0.958
    series_11_13[i]['<I>/A'] = series_11_13[i]['<I>/mA'] / 1000

    # Filter for cycle number
    x = series_11_13[i].loc[series_11_13[i]['cycle number'] == selected_cycle]['E vs RHE']
    y = series_11_13[i].loc[series_11_13[i]['cycle number'] == selected_cycle]['<I>/A']

    ax.scatter(x, y, label = labels[i], c='C00')

    # Check the middle region
    left_bound = .39
    right_bound = .50
    plt.vlines([left_bound, right_bound], -6.0e-4, 0, colors='C01')
    ilc_est.append(series_11_13[i].loc[series_11_13[i]['E vs RHE'].between(left_bound, right_bound)]['<I>/A'].mean())
    plt.scatter((left_bound+right_bound)/2, ilc_est[i], c='C03', marker='_', s=300)



    # ax.tick_params(axis='both', which='both', direction='in', right=True, top=True)
    ax.set_xlabel('$E$ vs. RHE / V')
    ax.set_ylabel('$i$ / A')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    fig.legend(loc=(0.75, 0.78))
    fig.tight_layout()
    plt.show()

print(np.array(ilc_est) * 1000)
```

```python
omega = np.array([1200, 1400, 1600])

x = (omega*2*np.pi / 60)**(1/2) # Convert to rad/s, then sqrt
y = np.array(ilc_est)

fit = PolyReg(x, y, 1)
print(fit.report())

fig, ax = plt.subplots()

ax.scatter(x, y * 1000, label='Data') # Plot in mA, fit in A
ax.plot(x, fit.eval(x) * 1000, label='Fit', color='C01') # Plot in mA, fit in A
ax.set_xlabel('$\omega^{-1/2}$ / rad$^{1/2}$ s$^{-1/2}$')
ax.set_ylabel('$i_{lc}$ / mA')
ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
# fig.legend(loc=(0.75, 0.78))
fig.tight_layout()
plt.show()
fig.savefig(os.path.join(dir_name, 'ilc-omega0.5.png'))

fit.report()
```
