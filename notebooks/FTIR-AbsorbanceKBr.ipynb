```python
import pandas as pd
import matplotlib.pyplot as plt
plt.close('all')
plt.style.use('JGW.mplstyle')
```

```python
# KBr_JGW-A-53-03
data = pd.read_csv(r'C:\Users\Administrator\OneDrive\Documents\PostBac\FTIR\KBr_JGW-A-53-03.csv', 
                   index_col=0, 
                   header=4, 
                   names=['Wavenumber', 'Absorbance'])


x, y = data.index, data['Absorbance']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
ax.invert_xaxis()
plt.grid()
ax.set_ylabel('Absorbance (a.u.)')
ax.set_xlabel('Wavenumber (cm$^{-1}$)')
ax.set_title('Pure KBr')
# plt.savefig('KBr_JGW-A-53-03')
```
